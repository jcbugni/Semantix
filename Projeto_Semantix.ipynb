```
#cria coluna com base se o candidato solicitou algum recurso especializado especifico
df['RECURSO_ESPECIFICO'] = np.where(
                 ((df['IN_SEM_RECURSO']==1) | (df['IN_BRAILLE']==1) | (df['IN_AMPLIADA_24']==1) | (df['IN_AMPLIADA_18']==1) | (df['IN_GUIA_INTERPRETE']==1) |
                 (df['IN_LEDOR']==1) | (df['IN_ACESSO']==1) | (df['IN_TRANSCRICAO']==1) | (df['IN_COMPUTADOR']==1) | (df['IN_CADEIRA_ESPECIAL']==1) |
                 (df['IN_LIBRAS']==1) | (df['IN_TEMPO_ADICIONAL']==1) | (df['IN_LEITURA_LABIAL']==1) | (df['IN_MESA_CADEIRA_RODAS']==1) | 
                 (df['IN_MESA_CADEIRA_SEPARADA']==1) | (df['IN_APOIO_PERNA']==1) | (df['IN_CADEIRA_CANHOTO']==1) | (df['IN_CADEIRA_ACOLCHOADA']==1) |
                 (df['IN_MEDIDOR_GLICOSE']==1) | (df['IN_MAQUINA_BRAILE']==1) | (df['IN_SOROBAN']==1) | (df['IN_MARCA_PASSO']==1) |
                 (df['IN_SONDA']==1) | (df['IN_MEDICAMENTOS']==1) | (df['IN_SALA_INDIVIDUAL']==1) | (df['IN_SALA_ESPECIAL']==1) |
                 (df['IN_SALA_ACOMPANHANTE']==1) | (df['IN_MOBILIARIO_ESPECIFICO']==1) | (df['IN_MATERIAL_ESPECIFICO']==1) | (df['IN_NOME_SOCIAL']==1) |
                 (df['IN_PROVA_DEITADO']==1) | (df['IN_MOBILIARIO_OBESO']==1) | (df['IN_LAMINA_OVERLAY']==1) | (df['IN_PROTETOR_AURICULAR']==1)), 
                 "Sim", "Não")
```

```
df = df.drop(columns=['IN_BAIXA_VISAO','IN_CEGUEIRA','IN_SURDEZ','IN_DEFICIENCIA_AUDITIVA','IN_SURDO_CEGUEIRA','IN_DEFICIENCIA_FISICA','IN_DEFICIENCIA_MENTAL',
                      'IN_DEFICIT_ATENCAO','IN_DISLEXIA','IN_DISCALCULIA','IN_AUTISMO','IN_VISAO_MONOCULAR','IN_OUTRA_DEF'])
```
